```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("Employee.csv")
```

```python
df.head(7)
```

```python
df.iloc[2:-2]
```

```python
# a) Add a new column 'color' based on the 'Department'
df['color'] = df['Department'].map({'IT': 'red', 'HR': 'green', 'Finance': 'blue', 'Marketing': 'orange'})
```

```python
# b) Create a simple histogram of the 'Department' column

plt.hist(df['Department'], color='skyblue', edgecolor='black', bins=len(df['Department'].unique()))
plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Histogram of Department')
plt.show()
```

```python
# Create a Lineplot of Age and Experience for different departments using Seaborn with default palette
plt.figure(figsize=(10, 6))

sns.lineplot(x='Age', y='Experience', hue='Department', data=df, markers=True)

plt.xlabel('Age')
plt.ylabel('Experience')
plt.title('Lineplot of Age and Experience for Different Departments')
plt.show()
```

```python
# Create a Lineplot of Age and Experience for different departments
plt.figure(figsize=(10, 6))

for department in df['Department'].unique():
    department_data = df[df['Department'] == department]
    plt.plot(department_data['Age'], department_data['Experience'], label=department, marker='o')

plt.xlabel('Age')
plt.ylabel('Experience')
plt.title('Lineplot of Age and Experience for Different Departments')
plt.legend()
plt.show()
```

```python
# d) Create histograms of Age for IT and HR departments in one figure
plt.hist([df[df['Department'] == 'IT']['Age'], df[df['Department'] == 'HR']['Age']], color=['red', 'green'], label=['IT', 'HR'])
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Age for IT and HR Departments')
plt.legend()
plt.show()
```

```python
# Create histograms of Age for IT and HR departments using Seaborn
plt.figure(figsize=(10, 6))

sns.histplot(data=df, x='Age', hue='Department', multiple='stack', bins=20, alpha=0.7)

plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Age for IT and HR Departments')
plt.legend()
plt.show()
```

```python
# Create histograms of Age for IT and HR departments in one figure
plt.figure(figsize=(10, 6))

# Histogram for IT
plt.hist(df[df['Department'] == 'IT']['Age'], bins=20, alpha=0.7, label='IT', color='red')

# Histogram for HR
plt.hist(df[df['Department'] == 'HR']['Age'], bins=20, alpha=0.7, label='HR', color='green')

plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Age for IT and HR Departments')
plt.legend()
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# Define the URL of the website
url = "http://www.values.com/inspirational-quotes"
```

```python
# Send an HTTP GET request to the URL
response = requests.get(url)
```

```python
# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all the elements with class "col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top"
    quote_elements = soup.find_all(class_='col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top')

    # Create a CSV file for storing the data
    with open('quotes_data.csv', 'w', newline='', encoding='utf-8') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['theme', 'url', 'Img','lines','author'])

        # Loop through each quote element
        for quote_element in quote_elements:
            # Extract theme
            theme = quote_element.find('h5').text.strip()
            # Extract URL if available
            url_element = quote_element.find('a')
            url = url_element['href'] if url_element else 'N/A'
            # Extract img URL if available
            img_element = quote_element.find('img')
            img_url = img_element['src'] if img_element else 'N/A'

            quote_lines = quote_element.find('img')['alt']
            author = quote_lines.split('#')[1].strip()
            
            # Write the data to the CSV file
            csv_writer.writerow([theme, url, img_url,quote_lines,author])

            # Print the extracted information
            print(f"Theme: {theme}")
            print(f"URL: {url}")
            print(f"Image URL: {img_url}")
            print(f"Lines: {quote_lines}")
            print(f"Author: {author}\n")


    print("Successfully scraped and saved to quotes_data.csv")
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

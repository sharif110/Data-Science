```python
import pandas as pd
 
# load the csv
data = pd.read_csv("nba.csv")
 
# show first 5 column
data.head()
```

```python
import seaborn as sns
```

```python
sns.distplot(data["Age"], color = "g")
```

```python
sns.distplot(data["Age"],bins=10,kde=False)
```

```python
sns.lineplot(x='Age', y='Weight', data=data)
```

```python
# plotting lineplot
sns.lineplot( data['Age'], data['Weight'])
```

```python
sns.scatterplot(x='Age', y='Weight', data=data)
```

```python
sns.scatterplot(data['Age'],data['Weight'])
```

```python
sns.scatterplot( x='Age', y='Weight', data=data, hue =data["Position"])
```

```python
sns.boxplot( data['Age'] )
```

```python
sns.boxplot( x='Age', y='Weight', data=data)
```

```python
sns.violinplot(data['Age'])
```

```python
sns.countplot(data["Position"])
```

```python
# Convert "Position" column to categorical
data["Position"] = pd.Categorical(data["Position"])

# Plot countplot
sns.countplot(x="Position", data=data)
```
